```python
def plotCubesAndPspecs(cubeList,pspecList):
    f = plt.figure(figsize=(15,6),dpi=400)
    f.subplots_adjust(hspace=0.2, wspace=0.2)
    
    ax1 = plt.subplot2grid((4,10),(0,0),rowspan=2,colspan=2)
    ax1 = plt.imshow(cubeList[0],cmap="Spectral")
    ax2 = plt.subplot2grid((4,10),(0,2),rowspan=2,colspan=2)
    ax2 = plt.imshow(cubeList[1],cmap="Spectral")
    ax3 = plt.subplot2grid((4,10),(2,0),rowspan=2,colspan=2)
    ax3 = plt.imshow(cubeList[2],cmap="Spectral")
    ax4 = plt.subplot2grid((4,10),(2,2),rowspan=2,colspan=2)
    ax4 = plt.imshow(cubeList[3],cmap="Spectral")
    
    for ax in f.axes:
        for t in ax.get_xticklabels()+ax.get_yticklabels():
            t.set_visible(False)
            
    ax5 = plt.subplot2grid((4,10),(0,5),rowspan=4,colspan=4)
    ax5 = plt.loglog(pspecList[0][:,0]/0.7,pspecList[0][:,1],color="Red",\
                     label="$z=11.5$, $x_{HI}=0.83$", ls='--',lw=2)
    ax5 = plt.loglog(pspecList[1][:,0]/0.7,pspecList[1][:,1],color="Orange",\
                     label="$z=10.5$, $x_{HI}=0.72$",lw=2)
    ax5 = plt.loglog(pspecList[2][:,0]/0.7,pspecList[2][:,1],color="Green",\
                     label="$z=9.5$, $x_{HI}=0.52$",ls='-.',lw=2)
    ax5 = plt.loglog(pspecList[3][:,0]/0.7,pspecList[3][:,1],color="Blue",\
                     label="$z=8.5$, $x_{HI}=0.20$",ls=":",lw=2)
    
    counter = 1
    for ax in f.axes:
        if counter == 5:
            ax.set_xlabel("$k$ [$h$Mpc$^{-1}\!$ ]",fontsize=20)
            ax.set_ylabel("$\Delta^2\!(k)$ [mK$^2\!$]",fontsize=20)
            ax.legend(loc="lower right", numpoints=1, fontsize=16,
              fancybox=True, borderpad=0.2, handlelength=2, labelspacing=0.1)
            ax.xaxis.set_tick_params(which='major',length=10)
            ax.xaxis.set_tick_params(which='minor',length=5)
            ax.yaxis.set_tick_params(which='major',length=10)
            ax.yaxis.set_tick_params(which='minor',length=5)
        elif counter == 1:
            ax.text(5,20, "$z=11.5$", bbox={"facecolor":"white"}, fontsize=16)
        elif counter == 2:
            ax.text(5,20, "$z=10.5$", bbox={"facecolor":"white"}, fontsize=16)
        elif counter == 3:
            ax.text(5,20, "$z=9.5$", bbox={"facecolor":"white"}, fontsize=16)
        elif counter == 4:
            ax.text(5,20, "$z=8.5$", bbox={"facecolor":"white"}, fontsize=16)
        counter +=1
        for t in ax.get_xticklabels()+ax.get_yticklabels():
            t.set_fontsize(18)

    savefig("cubesAndPspecs.png", bbox_inches='tight')
    return None
```

```python
plotCubesAndPspecs([temp5,temp4,temp3,temp2],[pspec5,pspec4,pspec3,pspec2])
```
